```python
#Created: 07Feb2022
#Author: Lynn Menchaca

#Resources
#Ken Jee - Data Science Project from Scratch series
#https://www.youtube.com/playlist?list=PL2zq7klxX5ASFejJj80ob9ZAnBHdz5O1t

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import cross_val_score
import statsmodels.api as sm

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
```

```python
raw_data_file_path = '/Users/lynnpowell/Documents/DS_Projects/DS_Salary_Project/Data_Science_Jobs_NoDescription.csv'
data_file_path = '/Users/lynnpowell/Documents/DS_Projects/DS_Salary_Project/Salary_Data_EDA.csv'
df_out = pd.read_csv(data_file_path)
df = df_out.drop(['Unnamed: 0'], axis = 1)
df.head()
```

```python
#choose relevant columns
#get dummy data for categorical data
#     dummy data column for each category and 1 as the dummy data when the categories match the new column
#     This will have implications for the type of model that we use
#train test split
#     typical sets: train, test, validation
#multiple linear regression
#lasso regression
#random forest (tree based model)
#could use: gradient 
#tune models using gridsearchCV
#test ensembles
```

```python
df.columns
```

```python
#choosing relavent columns
df_model = df[['Avg_Salary','Rating', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'Hourly', 
             'Employer_Provided', 'Job_State', 'Age', 'job_simp', 'seniority']]
```

```python
#gt dummy data
df_dum = pd.get_dummies(df_model)
df_dum.head()
```

```python
#dropping all na values
df_drop = df_dum.dropna()

#train test split
x = df_drop.drop('Avg_Salary', axis = 1)

# adding .values on the end makes it an array (better common practice)
# without .values -> it would be a series
y = df_drop.Avg_Salary.values

#20% in test set, 80% in train set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
#multiple linear regression
#using both statsmodel and sklearn model to compare
#For stats models have to create a constant (x)
#the constant creates the constant in the model
#Have to have a line and a intercept the constant is the intercept -> collumn of all 1s

#Statsmodels ols regression
x_sm = x = sm.add_constant(x)
model = sm.OLS(y,x_sm)
model.fit().summary()
```

```python
#looking at R-squared, P>|t| column is the focus
#P values of < 0.05 to 0.00 means it is significat in the model -> then look at the coef (coefficient)
#The coefficient is how much is added to the salary or how much is taken off (-values)

#R-squared seems low on 27%, so our model only explains 27% of the variation in the data
#Example with significant P values: 
#Hourly (paying less), Age (paying less/even), 
#Job_State_NY (paying more), Job_State_TX (paying less), 
#job_simp_analyst (paying less), job_simp_data scientist (paying more)

#Reasonable amount of multicollinearity so not putting to much weight in to the results
#More expoloratory to understand the data
#Good to use as a baseline to compare to some of the other models
```

```python
#sklearn model linear regression
#Doing the same thing except we are going to cross validate it

lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
#cross_val_score using neg mean absolute error
#This will be the baseline all other tests will be evaluated off of.
#This will show how far on average we are off from our general prediction
#So if we are off by 21 we are off by $21,000 average salary

cross_val_score(lm,x_train, y_train, scoring = 'neg_mean_absolute_error', cv = 3)
```

```python
#cross val score results
#after taking the mean, on averge we see we over by $33,213
#the matrix is sparse for a multiple linear regression
#difficult to get values becuase it is a limited data
#The lasso regression model will be able to normalize this
np.mean(cross_val_score(lm,x_train, y_train, scoring = 'neg_mean_absolute_error', cv = 3))
```

```python
#lasso regression
#normalization term alpha
#if alpha is zero it is the same as the multiple linear regression
#As we increase alpha it increases the amount the data is smoothed
#alpha default = 1

#alpha = 0.02 is discovered later
lm_l = Lasso(alpha=0.02)
lm_l.fit(x_train,y_train)
np.mean(cross_val_score(lm_l,x_train, y_train, scoring = 'neg_mean_absolute_error', cv = 3))
```

```python
#with alpha = 1 starts out worse being off by $36,047
#going to try multiple alphas to see which one is the best
alpha = []
error = []

for i in range(1,100):
    alpha.append(i/100)
    lml = Lasso(alpha=(i/100))
    error.append(np.mean(cross_val_score(lml,x_train, y_train, scoring = 'neg_mean_absolute_error', cv = 3)))

plt.plot(alpha, error)
```

```python
#zooming in closer to the peak
err = tuple(zip(alpha, error))
df_err = pd.DataFrame(err, columns = ['alpha', 'error'])
df_err[df_err.error == max(df_err.error)]
```

```python
#alpha 0.02 is giving us the best error term of -31.794 -> $31,794
#Improved a little bit from the cross validation of -33.213
```

```python
#Random Forest Regressor
#We expect random forest regressor to work well because we have a lot of 0s and 1s from our category values
#Also because it is a tree based decision process
#We donâ€™t have to worry at all about multicollinearity with this type of model

rf = RandomForestRegressor()
cross_val_score(rf, x_train, y_train, scoring = 'neg_mean_absolute_error', cv = 3)
```

```python
#My value improved (1 less), but not as much as Ken's (5 less then the lasso regression) in the tutorial
#I am surprised by these results
np.mean(cross_val_score(rf, x_train, y_train, scoring = 'neg_mean_absolute_error', cv = 3))
```

```python
#Going to tune the Random Forest Regressor using gridsearchCV
parameters = {'n_estimators':range(10,300,10), 'criterion':('mse','mae'),'max_features':('auto','sqrt','log2')}
gs = GridSearchCV(rf, parameters, scoring = 'neg_mean_absolute_error', cv = 3)
gs.fit(x_train, y_train)
```

```python
#It took about 40 min to compile the line above
gs.best_score_
gs.best_estimator_
```

```python
# test ensembles
tpred_lm = lm.predict(x_test)
#using alpha = 0.02
tpred_lml = lm_l.predict(x_test)
tpred_rf = gs.best_estimator_.predict(x_test)
```

```python
mean_absolute_error(y_test, tpred_lm)
```

```python
mean_absolute_error(y_test, tpred_lml)
```

```python
mean_absolute_error(y_test, tpred_rf)
```

```python
#random forest regressor has the best results
```

```python
#Combining different models to try and get better results
#trying to see if one model is over training
#this gave better results than the lasso model by its self, but not as good as the tuned random forest regressor
mean_absolute_error(y_test, (tpred_lm+tpred_rf)/2)
```
