```python
#columns to clean
#Salary Parsing
#Company Name text only
#State field
#See if job position is at the Headquaters
#age of company
#parsing of job description (python, etc.)
#Remove -1 in salary, see how many we have left
```

```python
#Removing -1 in Salary Estimate Column
df = df[df['Salary Estimate'] != '-1']
df.head()
```

```python
#Removing (Glassdoor est.) in Salary Estimate collumn
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary.head(20)
```

```python
#Remove the K and $ to leave just numbers
minus_Kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
minus_Kd.head(20)
```

```python
#In main data frame, making a column for per hour and employee provided salary
df['Hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['Employer_Provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'Employer Provided Salary:' in x.lower() else 0)
```

```python
#Removing the per hour and employer provided test from the Salary Estimate column
min_hr = minus_Kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
#Spliting the Salary Estimate column in to min and max salary columns and creating an averge collumn
df['Min_Salary'] = min_hr.apply(lambda x: int(x.split(' - ')[0]))
df['Min_Salary'].dtype
#df['Max Salary'] = min_hr.apply(lambda x: try int(x.split(' - ')[1]) except int(x.split(' - ')[0]))
#df['Max Salary'] = min_hr.apply(lambda x: int(x.split(' - ')[1]) if str(x.split(' - ')[1]).isdigit() else int(x.split(' - ')[0]))

def max_value(x):
    try:
        return int(x.split(' - ')[1])
    except:
        return int(x.split(' - ')[0])

df['Max_Salary'] = min_hr.apply(lambda x: max_value(x))
df['Avg_Salary'] = (df.Min_Salary+df.Max_Salary)/2
```

```python
#df.head(20)
```

```python
#Removing Rating in Company collumn
df['Company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-4], axis = 1)
df.Company_txt.head(10)
```

```python
#Sparateing the state the job is in
#df['Job_State'] = df['Location'].apply(lambda x: x.split(',')[1])

def state_split(x):
    try:
        return x.split(',')[1]
    except:
        return 'Remote'

df['Job_State'] = df['Location'].apply(lambda x: state_split(x))

df['Job_State'].value_counts()
#df.Job_State.value_counts()
```
